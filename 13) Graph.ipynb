```python
#whole dataframe is in histogram
df.hist(figsize = (20,10), color = 'g', alpha = 0.2) # alpha is the intesity of color
```

```python
#boxplot
df.plot(kind = 'box', figsize = (20,10))
```

```python
#vertical box plot
df.plot(kind = 'box' ,figsize = (20,10),color = {'boxes' : 'g' , 'whiskers' :'r'} , vert = False )
```

```python
#area Plot
df.plot(kind = "area", figsize =(20,10), alpha = 0.5, stacked = False)
```

```python
#Stacked = True,one on another kind of stacked chart. 
df.plot(kind = "area", figsize =(20,10), alpha = 0.5, stacked = True)
```

```python
#scatter plot, need 2 axis data
df.plot.scatter(x ='SepalLengthCm', y = 'PetalLengthCm' )
```

```python
df.plot.scatter(x ='SepalLengthCm', y = 'PetalWidthCm' )
```

```python
#scatter plot with 3 axis, c axis is a color axis
df.plot.scatter(x ='SepalLengthCm', y = 'PetalLengthCm', c = 'SepalWidthCm' )
```

```python
#s is a size of the petals width
df.plot.scatter(x ='SepalLengthCm', y = 'PetalLengthCm', c = 'SepalWidthCm', s = 600 )
```

```python
#s is defined with PetalWidthCm
df.plot.scatter(x ='SepalLengthCm', y = 'PetalLengthCm', c = 'SepalWidthCm', s = df['PetalWidthCm']*600 )
```
