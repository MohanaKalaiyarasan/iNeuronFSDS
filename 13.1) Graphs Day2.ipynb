```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris') #inbuilt dataset, name| we can load any dataset also
```

```python
df
```

```python
#scatter plot
sns.scatterplot(x=df.sepal_length, y=df.petal_length)
#seaborn default as x and y label
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df.sepal_length, df.petal_length, 'o') #``'o'`` circle marker
plt.xlabel('sep_len')
plt.ylabel('plt_len')
#matplotlib need to specify the labels
```

```python
#histogram
plt.hist(df['sepal_length'], alpha= 0.5) #alpha is a color contrast
```

```python
#density plot / distance plot
sns.distplot(df['sepal_length'])
```

```python
df1 = sns.load_dataset('tips') #another dataset Tips
```

```python
df1
```

```python
#relation plot
sns.relplot(x='total_bill', y='tip', data = df1)
```

```python
#to know how many of them are smoker or not
#1st method: not nice
df1.groupby('smoker').count()
```

```python
#2nd method: use value_counts
df1['smoker'].value_counts()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data=df1, hue='smoker') #hue is color group by
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data=df1, style = 'smoker')
```

```python
df1.head(2)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data=df1, style = 'size') #group by size
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data=df1, style = 'time', hue = 'smoker')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data=df1, size = 'size')
```

```python
sns.relplot(x='total_bill', y='tip', kind = 'line', data=df1)
```

```python
#create a graph based on column categories
sns.relplot(x='total_bill', y= 'tip', col='time', data=df1) #col - column groupby
```

```python
sns.relplot(x='total_bill', y= 'tip', col='size', data=df1)
```

```python
sns.relplot(x='total_bill', y= 'tip', col='sex', data=df1)
```

```python
df1
```

```python
sns.relplot(x='total_bill', y= 'tip', col='smoker', hue='sex', data=df1) 
```

```python
#categorial plot
sns.catplot(x='day', y='total_bill', data = df1)
```

```python
sns.catplot(x='day', y='total_bill', col='time', data = df1)
```

```python
sns.catplot(x='day', y='tip', data = df1)
```

```python
sns.catplot(x='day', y='tip', data = df1, kind='swarm') #this is not very useful
#pick the graph based on the requirement/utilization
```

```python
#boxplot
sns.catplot(x='day', y='tip', data = df1, kind='box',hue = 'sex')
```
