# FOR Loop with Else

```python
l = ["name", "age", "email", "address"]
```

```python
for i in l:
    print(i)
else: #else will follow once the above FOR loop completes
    print(" IF FOR LOOP able to complete then it come to ELSE")
```

```python
for i in l:
    if i == "name":
        break #it break and exit the loop, whatever below will not be executed.
    print(i)
else:
    print("check the statement")
    
#here there is no output as it is break the statement.
```

```python
for i in l:
    if i == "address":
        break
    print(i)
else:
    print("check the statement")
#here "break" will break the entire loop and exit from the loop.
#1st: if name==address(false), print>name
#2nd: if age==address(false), print>age
#3rd: if email==address(false), print>email
#4th: if address==address(True), now BREAK and exit from the loop. STOP.
```

```python
for i in l:
    if i == "address":
#         break
        print(i, "True")
    print(i, l.index(i))
else:
    print("check the statement")
#here "break" will break the entire loop and exit from the loop.
#1st: if name==address(false), print>name
#2nd: if age==address(false), print>age
#3rd: if email==address(false), print>email
#4th: if address==address(True), now NO BREAK and continue to print below items and Else statement also.
```

```python
a = 1
while a <5:
    print(a)
    if a == 3:
        continue
    a+=1
#continue, will print out continuously without going to next
# line 6 (here 6) and return to while loop at line 2
```

```python
#how to break the loop for above case
a = 1
while a <5:
    print(a)
    a+=1 #1st "a" increment
    if a == 3:
        continue
#     a+=1 #2nd "a" increment
#if "a" increment is at 2nd location(line8), when it reach a==3, it start to continue instead of going to line8,
#it jumps to line3 and start printing only 3 until we break at keyboard.
# by inserting "a" increment at 1st location(line5), upon reached a==3, it continue and goes to line3 and print 3 then
#increment to 4 and print 4 also then stop.
```
