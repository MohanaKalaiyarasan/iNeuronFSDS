```python
def filword(file_path):
    '''This function will help to filter out only the word file from a directory'''
    import os
    a = os.walk(file_path)#list down all subfolder as tree format
    l=[]
    dir_lst=list(a)
    
    import re
    pattern=re.compile(r"\.docx$")
    
    for lst in dir_lst:
        if type(lst)==tuple:
            for dirct in lst:
                if type(dirct)==list:
                    for file in dirct:
                        matches=pattern.finditer(file)
                        for match in matches:
                            l.append(file)
    return l,dir_lst

directory_path = input("Please enter the directory path to list .doc files in it: ")
doc_list,dir_list=filword(directory_path)
print("\nList of all .doc files are :\n", doc_list)
print("\n")
print("List of all files in the directory: \n", dir_list)
# C:\\Users\\mral\\Mohana Kalaiyarasan_Data Scientist              
```

```python
def appPDF(filename1, filename2, filename3):
    '''
    This function will be able to append two PDF files to a new file
    '''
    import PyPDF2 as pdf
    f1=open(filename1+'.pdf','rb')
    f2=open(filename2+'.pdf','rb')
    f3=open(filename3+'.pdf','wb')
    
    f1_read=pdf.PdfFileReader(f1)
    f2_read=pdf.PdfFileReader(f2)
    
    f_writer = pdf.PdfFileWriter()
    
    for page_no in range(f1_read.numPages):
        f1_pages =f1_read.getPage(page_no)
        f_writer.addPage(f1_pages)
    for page_no in range(f2_read.numPages):
        f2_pages =f2_read.getPage(page_no)
        f_writer.addPage(f2_pages)
    
    f_writer.write(f3)
    f1.close()
    f2.close()
    f3.close()
file_1 = input("Please enter the first pdf filename: ")    
file_2 = input("Please enter the second pdf filename: ")  
file_3 = input("Please enter the file to be created: ")  

appPDF(file_1, file_2, file_3)
#file1
#mywordfile
```

```python
#another 20 question:
def mergeTwoPDF(pdf1_path, pdf2_path, outfile_path):
    '''first 2 path is files, last one is outfile'''
    
    from PyPDF2 import PdfFileMerger
    files = [pdf1_path, pdf2_path]
    merger = PdfFileMerger()
    for pdf in files:
        merger.append(open(pdf,'rb'))
    with open(outfile_path, 'wb') as fileout:
        merger.write(fileout)
        
pdf1_path = "c:\\..pdf file1"
pdf2_path = "c:\\..pdf file2"
outfile_path ="c:\\..filename"

mergeTwoPDF(pdf1_path, pdf2_path, outfile_path)
```

```python
import cv2
def task11_video(video_path):
    video=cv2.VideoCapture(video_path)
    
    while(video.isOpened()):
        #capturing the video frame-by-frame
        ret,frame = video.read()
        if ret == True:
            cv2.imshow('Frame', frame) #showing the display of current frame
            if cv2.waitKey(25) & 0xFF == ord('x'): # Press x for exit the video
                break
        else:
            break
    video.release()
    cv2.destroyAllWindows()
        
```
